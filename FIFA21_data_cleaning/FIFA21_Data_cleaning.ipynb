```python
# Loading in required libraries
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# Reading in the FIFA 21  data
df_fifa = pd.read_csv('fifa21 raw data v2.csv')

# Taking a look at the first several winners
df_fifa.head(6)
```

```python
#Looking at all the columns this data set contains

print(df_fifa.columns)
```

```python
#Getting all the column information
print(df_fifa.info())
```

```python
#Checking if the dataset contains missing values or not 

print(df_fifa.isnull().sum())
```

```python
#dropping columns photoUrl and playerUrl 
df_fifa.drop(columns=df_fifa.columns[3:5], inplace=True)
df_fifa
```

```python
# As the data contains new-line characters hence replacing new line characters with empty strings

df_fifa=df_fifa.replace('\n',"" , regex=True)
```

```python
#In the height column, there are certain vlaues given in feet, hence converting all the values to cm

df = df_fifa.copy()

df['Height']=df['Height'].replace("cm", "", regex=True)
pd.set_option('display.max_rows', None)

def convert_height(height):
    if "'" in height:
        height = height[:-1]
        feet, inches = map(int, height.split("'"))
        total_inches = feet * 12 + inches
        return total_inches * 2.54
    else:
        return float(height)  # Already in centimeters

# Convert values in specified rows to centimeters
df['Height'] = df['Height'].apply(convert_height)

print(df['Height'].head(10))
    
```

```python
#converting value and wage to a float value and multiplying it by 1000000 (M) or 1000(K)

df['Value'] = df['Value'].replace(["€"], "", regex=True)

def convert_values(value):
    if "M" in value:
        val = value[:-1]
        return float(val)*1000000
    if "K" in value:
        val1 = value[:-1]
        return float(val1)*1000

# Convert values in specified rows to centimeters
df['Value'] = df['Value'].apply(convert_values)
print(df['Value'].head(10))
```

```python

df['Wage'] = df['Wage'].replace("€", "", regex=True)

def convert_to_money(amount):
    if "K" in amount:
        a = amount[:-1]
        return float(a)*1000
    else:
        float(amount)
df['Wage'] = df['Wage'].apply(convert_to_money)
```

```python
print(df['Wage'].head(10))
```

```python
#converting release claue to a float value and multiplying it by 1000000 (M) or 1000(K)

df['Release Clause'] = df['Release Clause'].replace("€", "", regex=True)
def convert_to_money2(amount):
    if "K" in amount:
        a = amount[:-1]
        return float(a)*1000
    if "M" in amount:
        a = amount[:-1]
        return float(a)*1000000
df['Release Clause'] = df['Release Clause'].apply(convert_to_money2)

```

```python
print(df['Release Clause'].head(10))
```

```python
#Removing the "★" from the "IR" column

df['IR'] = df['IR'].replace("★", "", regex=True).astype(int)
print(df['IR'].head(10))
```

```python
#Replacing all the NA values with 0

df_cleaned = df.fillna(0)
count = df_cleaned.isna().sum()
print(count)
```

```python
#Writing the cleaned data to a tsv file
df_cleaned.to_csv("cleaned_fifa", sep='\t')
```
